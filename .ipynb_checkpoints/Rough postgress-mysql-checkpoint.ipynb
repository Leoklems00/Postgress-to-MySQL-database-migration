```python
def extract_tables():
    cur.execute("select table_name \
    from information_schema.tables\
    where table_schema = 'public' and table_type = 'BASE TABLE';")
    cols = cur.fetchall()
    cols = [i[0] for i in cols]
    
    return cols
    
```

```python
tables = extract_tables()
tables
```

```python
def extract_fields(tables:list) -> dict:
    tables_dict = {}
    for table in tables:
        cur.execute(f"select column_name, udt_name, character_maximum_length \
        from information_schema.columns\
        where table_name = '{table}';")

        cols_dict = {}
        col_type = cur.fetchmany(10)
        cols = [i[0] for i in col_type]
        d_types = [i[1] for i in col_type]
        max_ls = [i[2] for i in col_type]
        for i in range(len(cols)):
            cols_dict[cols[i]] = [d_types[i], max_ls[i]]
        
        tables_dict[table] = cols_dict

    return tables_dict
    
```

```python
table_dicts = extract_fields(tables)
table_dicts
```
